**1) Exercício de herança múltipla e polimorfismo**
* Criar a classe **Veiculos** com um construtor que inclua marca e modelo
* Esta classe **Veículos** também terá de incluir um método que se chame **reabastecer()** o qual imprime por ecrã “Este veículo tem de reabastecer gasolina”
* Criar a classe **VEletricos** com um construtor que inclua marca, modelo e autonomia
* Esta classe VEletricos também terá de incluir um método que se chame **reabastecer()** o qual imprime por ecrã “Este veículo tem de reabastecer eletricidade”
* Criar a classe **BicicletaEletrica** que herde de **Veiculos** e de **VEletricos**, mas dando prioridade a **VEletricos** (já que é um veículo mais elétrico que normal)
* Criar a classe **Quadriciclo** que herde de **Veiculos** e de **VEletricos**, mas dando prioridade a **Veículos** (já que apenas usa a eletricidade de modo pontual)
* Criar um objeto de **BicicletaEletrica** e outro de **Quadriciclo**
* Empregar as técnicas de polimorfismo aprendidas para conseguir que ao perguntar-lhe aos objetos que acabamos de criar por reabastecimento, ambos nos respondam adequadamente


```python
# Completar o exercício aqui

class Veiculos:
    """
    Classe base que representa um veículo a combustão.

    Args:
        marca (str): A marca do veículo.
        modelo (str): O modelo do veículo.

    Atributos:
        marca (str): A marca do veículo.
        modelo (str): O modelo do veículo.
        
    Métodos:
        reabastecer: retorna impresso a mensagem de reabastecimento de gasolina.
    """
    def __init__(self, marca: str, modelo: str):
        self.marca = marca
        self.modelo = modelo

    def reabastecer(self):
        """
        Método que imprime a mensagem de reabastecimento de gasolina.
        """
        print("Este veículo tem de reabastecer gasolina")

class VEletrico:
    """
    Classe base que representa um veículo elétrico.

    Args:
        autonomia (int): A autonomia em quilômetros do veículo elétrico.

    Atributos:
        autonomia (int): A autonomia em quilômetros do veículo elétrico.
        
    Métodos:
        reabastecer: retorna impresso a mensagem de reabastecimento de eletricidade.
    """
    def __init__(self, autonomia: int):
        self.autonomia = autonomia

    def reabastecer(self):
        """
        Método que imprime a mensagem de reabastecimento de eletricidade.
        """
        print("Este veículo tem de reabastecer eletricidade")

class BicicletaEletrica(VEletrico, Veiculos):
    """
    Classe que representa uma bicicleta elétrica, herda de Veiculos e VEletrico.

    Args:
        marca (str): A marca da bicicleta elétrica.
        modelo (str): O modelo da bicicleta elétrica.
        autonomia (int): A autonomia em quilômetros da bicicleta elétrica.

    Atributos:
        marca (str): A marca da bicicleta elétrica.
        modelo (str): O modelo da bicicleta elétrica.
        autonomia (int): A autonomia em quilômetros da bicicleta elétrica.
    
     Métodos:
        reabastecer: retorna impresso a mensagem de reabastecimento de eletricidade, por herança.
    """
    def __init__(self, marca: str, modelo: str, autonomia: int):
        VEletrico.__init__(self, autonomia) # Call VEletrico constructor
        Veiculos.__init__(self, marca, modelo) # Call Veiculos constructor

class Quadriciclo(Veiculos, VEletrico):
    """
    Classe que representa um quadriciclo, herda de Veiculos e VEletrico.

    Args:
        marca (str): A marca do quadriciclo.
        modelo (str): O modelo do quadriciclo.
        autonomia (int): A autonomia em quilômetros do quadriciclo.

    Atributos:
        marca (str): A marca do quadriciclo.
        modelo (str): O modelo do quadriciclo.
        autonomia (int): A autonomia em quilômetros do quadriciclo.
        
    Métodos:
        reabastecer: retorna impresso a mensagem de reabastecimento de gasolina, por herança.
    """
    def __init__(self, marca: str, modelo: str, autonomia: int):
        Veiculos.__init__(self, marca, modelo)
        VEletrico.__init__(self, autonomia)

# Criar um objeto de BicicletaEletrica e outro de Quadriciclo
bicicleta_electrica = BicicletaEletrica("Trek", "Model X", 50)
quadriciclo = Quadriciclo("Polaris", "Quad 2000", 30)

# Usar polimorfismo para chamar o método reabastecer
print("Reabastecer a Bicicleta Elétrica:")
bicicleta_electrica.reabastecer()

print("\nReabastecer o Quadriciclo:")
quadriciclo.reabastecer()
```

**2) Exercício de herança**
* Criar a superclasse **Pessoal_Universitario** com um único atributo que seja um dicionário que inclua (Keys): “id”, “nome” e “email” 
* Criar as classes **Escritorio**, **Professor** e **Aluno**. Todas herdaram de **Pessoal_Universitário** e os seus dados
* No caso de **Escritorio**, deve adicionar ao seu dicionário de dados o dado “Gabinete”
* No caso de **Professor**, deve adicionar ao seu dicionário de dados o dado “Especializacao”
* No caso de **Aluno**, deve adicionar ao seu dicionário de dados o dado “CreditosAprovados" (integer)  
* Criar métodos que sirvam para mostrar a informação de cada classe
* Criar objetos de todas as classes
* Mostrar a informação desses objetos e verificar o seu output.

```python
# Completar o exercício aqui

class Pessoal_Universitario:
    """
    Classe base que representa o pessoal universitário.

    Args:
        id (int): O ID do pessoal universitário.
        nome (str): O nome do pessoal universitário.
        email (str): O email do pessoal universitário.

    Atributos:
        dados (dict): Um dicionário contendo os dados do pessoal universitário.

    Métodos:
        mostrar_informacao(): Exibe a informação do pessoal universitário.
    """
    def __init__(self, id: int, nome: str, email: str):
        self.dados = {
            "id": id,
            "nome": nome,
            "email": email
        }

    def mostrar_informacao(self):
        """
        Método que mostra a informação do pessoal universitário.
        """
        print("Informação do Pessoal Universitário:")
        for chave, valor in self.dados.items():
            print(f"\t{chave}: {valor}")


class Escritorio(Pessoal_Universitario):
    """
    Classe que representa o pessoal de escritório, herda de Pessoal_Universitário.

    Args:
        id (int): O ID do pessoal de escritório.
        nome (str): O nome do pessoal de escritório.
        email (str): O email do pessoal de escritório.
        gabinete (str): O número do gabinete do pessoal de escritório.

    Atributos:
        dados (dict): Um dicionário contendo os dados do pessoal de escritório.

    Métodos:
        mostrar_informacao(): Exibe a informação do pessoal de escritório.
    """
    def __init__(self, id, nome, email, gabinete):
        super().__init__(id, nome, email)
        self.dados["Gabinete"] = gabinete

    def mostrar_informacao(self):
        """
        Método que mostra a informação do pessoal de escritório.
        """
        #print("Informação do Pessoal de Escritório:")
        for chave, valor in self.dados.items():
            print(f"\t{chave}: {valor}")


class Professor(Pessoal_Universitario):
    """
    Classe que representa professores, herda de Pessoal_Universitário.

    Args:
        id (int): O ID do professor.
        nome (str): O nome do professor.
        email (str): O email do professor.
        especializacao (str): A especialização do professor.

    Atributos:
        dados (dict): Um dicionário contendo os dados do professor.

    Métodos:
        mostrar_informacao(): Exibe a informação do professor.
    """
    def __init__(self, id, nome, email, especializacao):
        super().__init__(id, nome, email)
        self.dados["Especializacao"] = especializacao

    def mostrar_informacao(self):
        """
        Método que mostra a informação do professor.
        """
        #print("Informação do Professor:")
        for chave, valor in self.dados.items():
            print(f"\t{chave}: {valor}")


class Aluno(Pessoal_Universitario):
    """
    Classe que representa alunos, herda de Pessoal_Universitário.

    Args:
        id (int): O ID do aluno.
        nome (str): O nome do aluno.
        email (str): O email do aluno.
        creditos_aprovados (int): O número de créditos aprovados pelo aluno.

    Atributos:
        dados (dict): Um dicionário contendo os dados do aluno.

    Métodos:
        mostrar_informacao(): Exibe a informação do aluno.
    """
    def __init__(self, id, nome, email, creditos_aprovados):
        super().__init__(id, nome, email)
        self.dados["CreditosAprovados"] = creditos_aprovados

    def mostrar_informacao(self):
        """
        Método que mostra a informação do aluno.
        """
        #print("Informação do Aluno:")
        for chave, valor in self.dados.items():
            print(f"\t{chave}: {valor}")


# Criar objetos de todas as classes
escritorio1 = Escritorio(1, "João", "joao@example.com", "Sala 101")
professor1 = Professor(2, "Maria", "maria@example.com", "Matemática")
aluno1 = Aluno(3, "Pedro", "pedro@example.com", 60)

escritorio2 = Escritorio(4, "Ana", "ana@example.com", "Sala 102")
professor2 = Professor(5, "Carlos", "carlos@example.com", "Física")
aluno2 = Aluno(6, "Sofia", "sofia@example.com", 45)

escritorio3 = Escritorio(7, "Mário", "mario@example.com", "Sala 103")
professor3 = Professor(8, "Laura", "laura@example.com", "História")
aluno3 = Aluno(9, "Rui", "rui@example.com", 75)

# Criar listas com os objetos
escritorios = [escritorio1, escritorio2, escritorio3]
professores = [professor1, professor2, professor3]
alunos = [aluno1, aluno2, aluno3]

# Rotina para apresentar os elementos de cada classe
def apresentar_elementos(classe, objetos):
    print(f"Informação dos {classe}:")
    for objeto in objetos:
        objeto.mostrar_informacao()
        print()

# Apresentar os elementos de cada classe
apresentar_elementos("Escritórios", escritorios)
apresentar_elementos("Professores", professores)
apresentar_elementos("Alunos", alunos)
```

**3)  Escolher a temática que preferir e definir uma estrutura de classes e subclasses onde pratique a herança em pelo menos dois níveis (herança multinível). Criar uma estrutura de mínimo 5 classes, com os seus respetivos construtores e métodos. Criar alguns objetos e comprovar que a herança esteja a funcionar. Se puder, incluir a técnica de polimorfismo e herança múltipla.**

```python
# Completar o exercício aqui

class Animal:
    """
    Classe base que representa um animal.

    Args:
        nome (str): O nome do animal.
        idade (int): A idade do animal.

    Atributos:
        nome (str): O nome do animal.
        idade (int): A idade do animal.

    Métodos:
        fazer_som: Retorna um som genérico de animal.
    """
    def __init__(self, nome: str, idade: int):
        self.nome = nome
        self.idade = idade

    def fazer_som(self):
        """
        Método que retorna um som genérico de animal.
        """
        return "Som genérico de animal"


class Mamifero(Animal):
    """
    Classe que representa um mamífero, herda de Animal.

    Args:
        nome (str): O nome do mamífero.
        idade (int): A idade do mamífero.
        pelos (str): A descrição dos pelos do mamífero.
        oviparo (bool): Indica se o mamífero é ovíparo.

    Atributos:
        pelos (str): A descrição dos pelos do mamífero.
        oviparo (bool): Indica se o mamífero é ovíparo.

    Métodos:
        fazer_som: Retorna um som genérico de mamífero.
    """
    def __init__(self, nome: str, idade: int, pelos: str, oviparo: bool):
        super().__init__(nome, idade)
        self.pelos = pelos
        self.oviparo = oviparo

    def fazer_som(self):
        """
        Método que retorna um som genérico de mamífero.
        """
        return "Som genérico de mamífero"


class Ave(Animal):
    """
    Classe que representa uma ave, herda de Animal.

    Args:
        nome (str): O nome da ave.
        idade (int): A idade da ave.
        envergadura_asas (str): A envergadura das asas da ave.
        migratoria (bool): Indica se a ave é migratória.

    Atributos:
        envergadura_asas (str): A envergadura das asas da ave.
        migratoria (bool): Indica se a ave é migratória.

    Métodos:
        fazer_som: Retorna um som genérico de ave.
    """
    def __init__(self, nome: str, idade: int, envergadura_asas: str, migratoria: bool):
        super().__init__(nome, idade)
        self.envergadura_asas = envergadura_asas
        self.migratoria = migratoria

    def fazer_som(self):
        """
        Método que retorna um som genérico de ave.
        """
        return "Som genérico de ave"


class Cachorro(Mamifero):
    """
    Classe que representa um cachorro, herda de Mamífero.

    Args:
        nome (str): O nome do cachorro.
        idade (int): A idade do cachorro.
        pelos (str): A descrição dos pelos do cachorro.
        oviparo (bool): Indica se o cachorro é ovíparo.

    Atributos:
        Nenhum atributo adicional.

    Métodos:
        fazer_som: Retorna um som de latido de cachorro.
        brincar: Retorna uma mensagem indicando que o cachorro está brincando.
    """
    def __init__(self, nome: str, idade: int, pelos: str, oviparo: bool):
        super().__init__(nome, idade, pelos, oviparo)

    def fazer_som(self):
        """
        Método que retorna um som de latido de cachorro.
        """
        return "Latido: Au Au!"

    def brincar(self):
        """
        Método que retorna uma mensagem indicando que o cachorro está brincando.
        """
        return f"{self.nome} está brincando."


class Passaro(Ave):
    """
    Classe que representa um pássaro, herda de Ave.

    Args:
        nome (str): O nome do pássaro.
        idade (int): A idade do pássaro.
        envergadura_asas (str): A envergadura das asas do pássaro.
        migratoria (bool): Indica se o pássaro é migratório.

    Atributos:
        Nenhum atributo adicional.

    Métodos:
        fazer_som: Retorna um som de canto de pássaro.
        voar: Retorna uma mensagem indicando que o pássaro está voando.
    """
    def __init__(self, nome: str, idade: int, envergadura_asas: str, migratoria: bool):
        super().__init__(nome, idade, envergadura_asas, migratoria)

    def fazer_som(self):
        """
        Método que retorna um som de canto de pássaro.
        """
        return "Canto: Trililililili!"

    def voar(self):
        """
        Método que retorna uma mensagem indicando que o pássaro está voando.
        """
        return f"{self.nome} está voando."
    
# Criar objetos
animal_generico = Animal("Animal Genérico", 5)
mamifero_generico = Mamifero("Mamífero Genérico", 3, "Marrom", False)
ave_generica = Ave("Ave Genérica", 2, "40 cm", True)
cachorro = Cachorro("Fido", 2, "Marrom", False)
passaro = Passaro("Piu", 1, "30 cm", False)

# Usar polimorfismo
animais = [animal_generico, mamifero_generico, ave_generica, cachorro, passaro]
for animal in animais:
    print(f"{animal.nome}: {animal.fazer_som()}")

# Herança múltipla
class AnimalDomestico(Animal):
    """
    Classe que representa um animal doméstico, herdando da classe Animal.

    Args:
        nome (str): O nome do animal doméstico.
        idade (int): A idade do animal doméstico.

    Atributos:
        nome (str): O nome do animal doméstico.
        idade (int): A idade do animal doméstico.

    Métodos:
        fazer_som(): Retorna um som genérico de animal doméstico.
    """

    def __init__(self, nome: str, idade: int):
        super().__init__(nome, idade)

    def fazer_som(self):
        """
        Método que retorna um som genérico de animal doméstico.
        """
        return "Som genérico de animal doméstico"


class CachorroDomestico(Cachorro, AnimalDomestico):
    """
    Classe que representa um cachorro doméstico, herdando de Cachorro e AnimalDomestico.

    Args:
        nome (str): O nome do cachorro.
        idade (int): A idade do cachorro.
        pelos (str): O tipo de pelos do cachorro.
        oviparo (bool): Indica se o cachorro é ovíparo.

    Atributos:
        nome (str): O nome do cachorro.
        idade (int): A idade do cachorro.
        pelos (str): O tipo de pelos do cachorro.
        oviparo (bool): Indica se o cachorro é ovíparo.

    Métodos:
        fazer_som(): Retorna um latido amigável.
    """

    def __init__(self, nome: str, idade: int, pelos: str, oviparo: bool):
        super().__init__(nome, idade, pelos, oviparo)

    def fazer_som(self):
        """
        Método que retorna um latido amigável.
        """
        return "Latido amigável: Au Au!"

cachorro_domestico = CachorroDomestico("Rex", 3, "Marrom", False)
print(f"{cachorro_domestico.nome}: {cachorro_domestico.fazer_som()}")
```
