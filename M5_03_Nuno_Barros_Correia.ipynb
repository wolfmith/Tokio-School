# JSON (JavaScript Object Notation)

1) Tenha em consideração o seguinte JSON String:

{"chefe_projeto":{"Nome":"Juan","Idade":28,"Experiência": ["Gestão","Finanças","Basesdedados"],"Residência": 
"Madrid","HorasProjeto":3500},"funcionários":[{"Nome":"Elena","Idade":26,"Experiência":
["JavaScript","Python"],"Residência":"Madrid","HorasProjeto":500},{"Nome":"Luis","Idade":31,"Experiência": 
["Django","Flask","Pyramid"],"Residência":"Barcelona","HorasProjeto":1100}]}


* Se necessário, utilizar um visualizador de JSON para o entender.
* O objetivo é criar um dicionário com todos os dados e estruturas internas necessárias para que seja igual ao JSON String.
* Criar estruturas mais pequenas até que se chegue a formar o mesmo JSON string que vemos acima.
* Armazenar esta estrutura (o dicionário) numa variável.
* Comprovar o seu tipo e mostrar por ecrã.
* Criar duas variáveis: horas_funcionários e horas_chefe
* Extrair as horas do chefe e introduzir na sua variável.
* Criar um ciclo que percorra a totalidade de funcionários que tem na estrutura e para cada funcionário somar as suas horas na sua variável correspondente.
* Somar todas as horas e imprimir.

```python
# Completa o exercicio aqui
import json

# String JSON fornecida
json_string = '{"chefe_projeto":{"Nome":"Juan","Idade":28,"Experiência": ["Gestão","Finanças","Basesdedados"],"Residência": "Madrid","HorasProjeto":3500},"funcionários":[{"Nome":"Elena","Idade":26,"Experiência": ["JavaScript","Python"],"Residência":"Madrid","HorasProjeto":500},{"Nome":"Luis","Idade":31,"Experiência": ["Django","Flask","Pyramid"],"Residência":"Barcelona","HorasProjeto":1100}]}'

# Converter a string JSON em um dicionário Python
dados = json.loads(json_string)
#print(dados)

# Comprovar o tipo da variável
print("Tipo da variável 'dados':", type(dados))

# Criar variáveis para horas do chefe e dos funcionários
horas_chefe = dados["chefe_projeto"]["HorasProjeto"]
horas_funcionários = 0

# Somar as horas de todos os funcionários
for funcionário in dados["funcionários"]:
    horas_funcionários += funcionário["HorasProjeto"]

# Somar todas as horas
total_horas = horas_chefe + horas_funcionários

# Imprimir o total de horas
print("Total de horas:", total_horas)
```

2) No exercício anterior trabalhou-se com um dicionário que nós mesmos/as criamos. Converter para um formato JSON String, imprimir o seu tipo e os seus dados. Para que serve este formato?

```python
# Completa o exercicio aqui

"""
O formato JSON (JavaScript Object Notation) é uma forma de representar dados organizados numa sequência de texto. 
É muito utilizado para a transferência de informações entre sistemas, porque é simples de ler e escrever para pessoas 
e fácil de interpretar por máquinas.

O resultado é uma cadeia de texto com formato JSON que pode ser usada para guardar informações, 
enviá-las através da rede ou partilhá-las com outros sistemas que compreendam JSON. 
É uma forma eficaz e padrão de representar dados estruturados.
"""

import json

# Converter o dicionário em JSON String
json_string = json.dumps(dados, indent=4, ensure_ascii=False)

# Imprimir o tipo do JSON String e seus dados
print("Tipo do JSON String:", type(json_string))
print("JSON String:")
print(json_string)

```

3) Por vezes encontramos JSON que se tem de modificar. Para isso, temos de descodificá-los, fazer as modificações pertinentes e voltar a codificá-los para os deixar como JSON de novo. No seguinte exemplo podemos ver que há alguns erros:

* Ao Superman falta-lhe como poder "Voar"
* No Batman, a idade é 35, não 350
* No Batman, retire-lhe o poder de "Raio laser"
* No Wonder Woman falta-lhe o poder "Laço da verdade"

Depois de corrigir, sem criar um novo, voltar a transformar estes dados num JSON String

```python
# NÃO UTILIZE ESTE BLOCO

superherois = {
	"nomedaEquipa": "Super Hero Squad",
	"cidade": "Metro City",
	"formado": 2016,
	"baseSecreta": "Super Tower",
	"activo": "Sim",
	"membros": [
		{
			"nome": "SuperMan",
			"idade": 29,
			"identidadeSecreta": "Clart Kent",
			"poderes": [
				"Super força",
				"Super velocidade",
				"Raio laser"
			]
		},
		{
			"nome": "Batman",
			"idade": 350,
			"identidadeSecreta": "Bruce Wayne",
			"poderes": [
				"Detective",
				"Dinheiro",
				"Raio laser"
			]
		},
		{
			"nome": "Wonder Woman",
			"idade": 900,
			"identidadeSecreta": "Diana de Temiscira",
			"poderes": [
				"Super força",
				"Super velocidade"
			]
		}
	]
}
```

```python
# Completa o exercicio aqui

import json

# Corrigir os erros nos dados dos super-heróis
superherois["membros"][0]["poderes"].append("Voar")  # Adicionar o poder "Voar" ao SuperMan
superherois["membros"][1]["idade"] = 35  # Corrigir a idade do Batman
superherois["membros"][1]["poderes"].remove("Raio laser")  # Remover o poder "Raio laser" do Batman
superherois["membros"][2]["poderes"].append("Laço da verdade")  # Adicionar o poder "Laço da verdade" à Wonder Woman

# Transformar os dados corrigidos em um JSON String
superherois_json = json.dumps(superherois, indent=4, ensure_ascii=False)

# Imprimir o JSON String corrigido
print(superherois_json)
```
