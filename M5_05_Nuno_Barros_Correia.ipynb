**1) Trabalhar com NumPy**

* Importar numpy e usar um alias para poder utilizar os seus recursos 
* Criar um array de elementos (de 10 a 100)
* Inverter o array
* Criar um array 3x3 com valores de 0 a 8 
* Criar um array identidade de 6x6
* Criar um array com valores aleatórios com o formato 3x3x3
* Criar dois arrays, com valores aleatórios, A e B, e verificar se são iguais (há várias opções, investigar)

```python
# Completar o exercício aqui

import numpy as np  # Import NumPy and alias it as np

# Create an array of elements from 10 to 100
array_10_to_100 = np.arange(10, 101)

# Invert the array
inverted_array = np.flip(array_10_to_100)

# Create a 3x3 array with values from 0 to 8
array_3x3 = np.arange(9).reshape(3, 3)

# Create a 6x6 identity array
identity_array = np.eye(6)

# Create a 3x3x3 array with random values
random_array = np.random.rand(3, 3, 3)

# Create two random arrays, A and B
A = np.random.rand(3, 3)
B = np.random.rand(3, 3)

# Check if arrays A and B are equal (element-wise)
are_equal = np.array_equal(A, B)

print("Array from 10 to 100:")
print(array_10_to_100)
print("\nInverted Array:")
print(inverted_array)
print("\n3x3 Array:")
print(array_3x3)
print("\n6x6 Identity Array:")
print(identity_array)
print("\n3x3x3 Random Array:")
print(random_array)
print("\nArrays A and B are equal:", are_equal)
```

**2) Seguir com NumPy**

* Criar um array de 100 elementos aleatórios compreendidos entre 1 e 1000 (podem ser números inteiros). 
* Este array irá chamar-se dólares
* Fazer a conversão dos preços em dólares do primeiro array para um segundo array que se vai chamar euros. Devemos utilizar a  seguinte conversão: 
  - 1 USD = 0,874413 EUR
* Mostrar o resultado do array de euros, arredondando a 2 casas decimais.

```python
# Completar o exercício aqui
import numpy as np

# Create an array of 100 random elements between 1 and 1000
dolares = np.random.randint(1, 1001, 100)
print("Array of Prices in Dollars:")
print(dolares)

# Conversion rate
conversion_rate = 0.874413

# Convert dollars to euros
euros = dolares * conversion_rate

# Round the result to 2 decimal places
euros = np.round(euros, 2)

# Display the resulting array of euros
print("\nArray of Prices in Euros:")
print(euros)
```

```python
# Completar o exercício aqui

# Import pandas with an alias
import pandas as pd

# Set the maximum width to a larger value to avoid splitting the print
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 1000)

# Códigos para formatar o texto dos prints
class color:
    PURPLE = '\033[95m'
    CYAN = '\033[96m'
    DARKCYAN = '\033[36m'
    BLUE = '\033[94m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    WHITE = '\033[97m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    END = '\033[0m'
    REDBACK = '\033[41m'
    REVERSE = '\u001b[7m'

# Load the CSV files into DataFrames
titles_df = pd.read_csv('05_05_imdb_titulos.csv')
elenco_df = pd.read_csv('05_05_imdb_elenco.csv')

# Show the first 5 records of each DataFrame
print(color.REVERSE + color.BOLD + "First 5 records of titles_df:" + color.END)
print(titles_df.head())
print()
print(color.REVERSE + color.BOLD + "First 5 records of elenco_df:" + color.END)
print(elenco_df.head())

# Show the number of records in each DataFrame
print(color.REVERSE + color.BOLD + "\nNumber of records in titles_df:" + color.END, len(titles_df))
print(color.REVERSE + color.BOLD + "Number of records in elenco_df:" + color.END, len(elenco_df))

# Show the 5 oldest titles
oldest_titles = titles_df.sort_values(by='year').head()
print(color.REVERSE + color.BOLD + "\n5 oldest titles:" + color.END)
print(oldest_titles)

# Count the number of movies with 'Dracula' in the title
dracula_movies = titles_df[titles_df['title'].str.contains('Dracula', case=False, na=False)]
print(color.REVERSE + color.BOLD + "\nNumber of movies with 'Dracula' in the title:" + color.END, len(dracula_movies))

# Show the 10 most common titles
common_titles = titles_df['title'].value_counts().head(10)
print(color.REVERSE + color.BOLD + "\n10 most common titles:" + color.END)
print(common_titles.to_string())

# Find the first movie titled "Romeo and Juliet"
# Filter the DataFrame to get movies titled 'Romeo and Juliet'
romeo_and_juliet_movies = titles_df[titles_df['title'] == 'Romeo and Juliet']
# Sort the filtered DataFrame by the 'year' column
romeo_and_juliet_movie = romeo_and_juliet_movies.sort_values(by='year').head(1)
print(color.REVERSE + color.BOLD + "\nFirst movie titled 'Romeo and Juliet':" + color.END)
print(romeo_and_juliet_movie.to_string())

# List movies with 'Exorcist' in the title, sorted by year
exorcist_movies = titles_df[titles_df['title'].str.contains('Exorcist', case=False, na=False)].sort_values(by='year')
print(color.REVERSE + color.BOLD + "\nMovies with 'Exorcist' in the title, sorted by year:" + color.END)
print(exorcist_movies.to_string())

# Count the number of movies made in 1950
movies_1950 = len(titles_df[titles_df['year'] == 1950])
print(color.REVERSE + color.BOLD + "\nNumber of movies made in 1950:" + color.END, movies_1950)

# Count the number of movies made from 1950 to 1959 (inclusive)
movies_1950s = len(titles_df[(titles_df['year'] >= 1950) & (titles_df['year'] <= 1959)])
print(color.REVERSE + color.BOLD + "\nNumber of movies made from 1950 to 1959 (inclusive):" + color.END, movies_1950s)

# Group and count roles in 'The Godfather' trilogy
godfather_roles = elenco_df[elenco_df['title'].str.contains('The Godfather', case=False, na=False)]
role_counts = godfather_roles.groupby('character')['character'].count()
print(color.REVERSE + color.BOLD + "\nRoles in 'The Godfather' trilogy and their counts:" + color.END)
print(role_counts.to_string())

# Show the complete cast of 'Dracula' (1958), sorted by 'n'
dracula_1958_cast = elenco_df[(elenco_df['title'] == 'Dracula') & (elenco_df['year'] == 1958)].sort_values(by='n')
print(color.REVERSE + color.BOLD + "\nComplete cast of 'Dracula' (1958), sorted by 'n':" + color.END)
print(dracula_1958_cast)

# Count the number of roles for 'Bruce Wayne' as a protagonist
bruce_wayne_roles = elenco_df[
    elenco_df['character'].str.contains(r'(?i)^(?!waiter).*Bruce Wayne', case=False, na=False) 
    & (elenco_df['type'] == 'actor')
]
# As instruções são ambíguas, por um lado pedem protagonistas, mas pelo outro dizem que a personagem secundária Young Bruce Wayne também deve ser incluída. A linha abaixo deverá ser incluída para os papeis como protagonista.
# protagonist_roles = bruce_wayne_roles[bruce_wayne_roles['n'] == 1]
print(color.REVERSE + color.BOLD + "\nNumber of roles for 'Bruce Wayne' as a protagonist:" + color.END, len(bruce_wayne_roles)) # Usar a variável protagonis_roles se pretender protagonistas

# Count the total number of roles for 'Robert de Niro'
robert_de_niro_roles = elenco_df[elenco_df['name'] == 'Robert De Niro']
print(color.REVERSE + color.BOLD + "\nTotal number of roles for 'Robert de Niro':" + color.END, len(robert_de_niro_roles))

# List protagonist roles for 'Charlton Heston' in the 60s, sorted by year
charlton_heston_60s_roles = elenco_df[(elenco_df['name'] == 'Charlton Heston') & (elenco_df['type'] == 'actor') & (elenco_df['year'] >= 1960) & (elenco_df['year'] <= 1969)].sort_values(by='year', ascending=False)
print(color.REVERSE + color.BOLD + "\nProtagonist roles for 'Charlton Heston' in the 60s, sorted by year:" + color.END)
print(charlton_heston_60s_roles)

# Count the number of roles for actors in the 50s
actors_50s_roles = len(elenco_df[(elenco_df['type'] == 'actor') & (elenco_df['year'] >= 1950) & (elenco_df['year'] <= 1959)])
print(color.REVERSE + color.BOLD + "\nNumber of roles for actors in the 50s:" + color.END, actors_50s_roles)

# Count the number of roles for actresses in the 50s
actresses_50s_roles = len(elenco_df[(elenco_df['type'] == 'actress') & (elenco_df['year'] >= 1950) & (elenco_df['year'] <= 1959)])
print(color.REVERSE + color.BOLD + "\nNumber of roles for actresses in the 50s:" + color.END, actresses_50s_roles)
```
