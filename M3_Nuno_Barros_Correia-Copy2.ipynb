````python
# Fazer o exercício aqui

# Definir a classe Aluno
class Aluno:
    """
    Represents a student with a name and a grade.

    Args:
        nome (str): The name of the student.
        nota (float): The grade of the student.

    Attributes:
        nome (str): The name of the student.
        nota (float): The grade of the student.

    Methods:
        qualificacao: Determines if the student passed or failed based on their grade.
        __str__: Returns a formatted string with the student's name and grade.
        
    """

    def __init__(self, nome, nota):
        """
        Initializes a new student with a name and a grade.

        Args:
            nome (str): The name of the student.
            nota (float): The grade of the student.

        """
        self.nome = nome
        self.nota = nota
        # Imprimir uma mensagem informando que o aluno foi criado com sucesso
        print(f"O aluno {self.nome} foi criado com sucesso.")

    def qualificacao(self):
        """
        Determines if the student passed or failed based on their grade.

        Returns:
            str: A message indicating if the student passed or failed.

        """
        if self.nota >= 5:
            return f"{self.nome} passou!"
        else:
            return f"{self.nome} reprovou."
        
    def __str__(self):
        """
        Returns a formatted string with the student's information.

        Returns:
            str: A formatted string with the student's name and grade.

        """
        return f"Nome: {self.nome}\nNota: {self.nota}"

# Create student objects
aluno1 = Aluno("Alice", 7)
aluno2 = Aluno("Bob", 4)
aluno3 = Aluno("Charlie", 9)

# Print student information and qualification results
print("\nInformações dos Alunos:")
print(aluno1)
print(aluno1.qualificacao())

print()
print(aluno2)
print(aluno2.qualificacao())

print()
print(aluno3)
print(aluno3.qualificacao())
````

No exercício anterior, ao fazer um print as notas e nomes, vamos obter a nota e nome do aluno. Como pode ser considerado um problema de segurança, vamos encapsular o código para que não se possa aceder diretamente às variáveis e o acesso tenha de ser através de métodos.

````python
# Realizar o exercício aqui
class Aluno:
    """
    Represents a student with a name and a grade.

    Args:
        nome (str): The name of the student.
        nota (float): The grade of the student.

    Attributes:
        __nome (str): The name of the student (private).
        __nota (float): The grade of the student (private).

    Methods:
        qualificacao: Determines if the student passed or failed based on their grade.
        get_nome: Gets the name of the student.
        set_nome: Sets the name of the student.
        get_nota: Gets the grade of the student.
        set_nota: Sets the grade of the student.
        __str__: Returns a formatted string with the student's name and grade.
        
    """

    def __init__(self, nome, nota):
        """
        Initializes a new student with a name and a grade.

        Args:
            nome (str): The name of the student.
            nota (float): The grade of the student.

        """
        self.__nome = nome
        self.__nota = nota
        print(f"O aluno {self.__nome} foi criado com sucesso.")

    def qualificacao(self):
        """
        Determines if the student passed or failed based on their grade.

        Returns:
            str: A message indicating if the student passed or failed.

        """
        if self.__nota >= 5:
            return f"{self.__nome} passou!"
        else:
            return f"{self.__nome} reprovou."

    def get_nome(self):
        """
        Gets the name of the student.

        Returns:
            str: The name of the student.

        """
        return self.__nome

    def set_nome(self, novo_nome):
        """
        Sets the name of the student.

        Args:
            novo_nome (str): The new name of the student.

        """
        self.__nome = novo_nome

    def get_nota(self):
        """
        Gets the grade of the student.

        Returns:
            float: The grade of the student.

        """
        return self.__nota

    def set_nota(self, nova_nota):
        """
        Sets the grade of the student.

        Args:
            nova_nota (float): The new grade of the student.

        """
        self.__nota = nova_nota

    def __str__(self):
        """
        Returns a formatted string with the student's information.

        Returns:
            str: A formatted string with the student's name and grade.

        """
        return f"Nome: {self.__nome}\nNota: {self.__nota}"

# Create student objects
aluno1 = Aluno("Alice", 7)
aluno2 = Aluno("Bob", 4)
aluno3 = Aluno("Charlie", 9)

# Attempt to directly access private variables (should raise an error)
try:
    print(aluno1.__nome)  # This should raise an AttributeError
except AttributeError:
    try:
        print(aluno1.nome)    # This should raise an AttributeError
    except AttributeError:
        print("\nNão é possível aceder às variáveis do objeto diretamente sem a utilização do método getter.")
        
# Attempt to directly modify private variables (should raise an error)
aluno2.__nota = 6     # This should not affect the grade of aluno2
aluno2.nota = 6       # This should not affect the grade of aluno2
print(f"\nNão foi possível modificar a nota do aluno2 sem a utilização do método setter, a nota é {aluno2.get_nota()}.")

# Print student information and qualification results using the get methods:

print("\nInformações dos Alunos:")
print(f"\tNome: {aluno1.get_nome()}")
print(f"\tNota: {aluno1.get_nota()}")
print(f"\t{aluno1.qualificacao()}")

print()
print(f"\tNome: {aluno2.get_nome()}")
print(f"\tNota: {aluno2.get_nota()}")
print(f"\t{aluno2.qualificacao()}")

print()
print(f"\tNome: {aluno3.get_nome()}")
print(f"\tNota: {aluno3.get_nota()}")
print(f"\t{aluno3.qualificacao()}")
````

### O plano cartesiano

Representa um espaço bidimensional (em 2 dimensões), formado por duas linhas perpendiculares, uma horizontal e outra vertical, que se intersetam num ponto. A linha horizontal é chamada eixo das abcissas ou __eixo X__, enquanto a vertical é chamada eixo das ordenadas ou simplesmente o __eixo Y__.
Quanto ao ponto de interseção, é conhecido como o ponto de origem O. __O plano está dividido em 4 quadrantes.__


<img src="./eixos.png"  />



### Pontos e coordenadas

O objetivo é descrever a posição dos **pontos** no plano sob a forma de **coordenadas**, que são formadas associando o valor do eixo X (horizontal) com o valor do eixo Y (vertical).


A representação de um ponto é simples: **P(X,Y)** onde X e Y são a distância horizontal (esquerda ou direita) e vertical (para cima ou para baixo) 
respetivamente, usando como referência o ponto de origem (0,0), mesmo no centro do plano.

<img src="./Puntos.png"  />


### Vetores no plano

Finalmente, um vetor no plano refere-se a um segmento orientado, gerado a partir de dois pontos diferentes.

Para efeitos práticos, é simplesmente uma linha formada a partir de um ponto inicial na direção de outro ponto final, pelo que se entende que um vetor tem comprimento e direção/sentido.

<img src="./Vector.png"  />


Nesta figura, podemos ver dois pontos A e B que poderíamos definir como se segue:
* **A(x1, y1)** => **A(2, 3)**
* **B(x2, y2)** => **B(7, 5)**

E o vetor seria representado como a diferença entre as coordenadas do segundo ponto em relação ao primeiro (o segundo menos o primeiro):
* **AB = (x2-x1, y2-y1)** => **(7-2, 5-3)** => **(5,2)** 

O que significa:

5 à direita e 2 para cima.

````python
# Fazer o exercício aqui

import math

class Ponto:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def quadrante(self):
        if self.x > 0 and self.y > 0:
            return "Primeiro Quadrante"
        elif self.x < 0 and self.y > 0:
            return "Segundo Quadrante"
        elif self.x < 0 and self.y < 0:
            return "Terceiro Quadrante"
        elif self.x > 0 and self.y < 0:
            return "Quarto Quadrante"
        elif self.x == 0 and self.y == 0:
            return "Origem"
        else:
            return "Sobre um eixo"

    def vetor(self, outro_ponto):
        vetor_x = outro_ponto.x - self.x
        vetor_y = outro_ponto.y - self.y
        return Ponto(vetor_x, vetor_y)

    def distancia(self, outro_ponto):
        distancia = math.sqrt((outro_ponto.x - self.x) ** 2 + (outro_ponto.y - self.y) ** 2)
        return distancia
    
    def __str__(self):
        return f"({self.x},{self.y})"

# Agora, criamos os pontos A(2, 3), B(5,5), C(-3, -1) e D(0,0)
ponto_A = Ponto(2, 3)
ponto_B = Ponto(5, 5)
ponto_C = Ponto(-3, -1)
ponto_D = Ponto(0, 0)

# Imprimimos os pontos
print(f"Ponto A: {ponto_A}")
print(f"Ponto B: {ponto_B}")
print(f"Ponto C: {ponto_C}")
print(f"Ponto D: {ponto_D}")

# Consultamos a que quadrantes pertencem os pontos A, C e D
print(f"A pertence ao {ponto_A.quadrante()}")
print(f"C pertence ao {ponto_C.quadrante()}")
print(f"D pertence ao {ponto_D.quadrante()}")

# Consultamos os vetores AB e BA
vetor_AB = ponto_A.vetor(ponto_B)
vetor_BA = ponto_B.vetor(ponto_A)
print(f"Vetor AB: {vetor_AB}")
print(f"Vetor BA: {vetor_BA}")

# Consultamos a distância entre os pontos 'A e B' e 'B e A'
dist_AB = ponto_A.distancia(ponto_B)
dist_BA = ponto_B.distancia(ponto_A)
print(f"Distância entre A e B: {dist_AB}")
print(f"Distância entre B e A: {dist_BA}")

# Determinamos qual dos 3 pontos A, B ou C está mais afastado da origem (0,0)
distancia_A = ponto_A.distancia(Ponto(0, 0))
distancia_B = ponto_B.distancia(Ponto(0, 0))
distancia_C = ponto_C.distancia(Ponto(0, 0))

if distancia_A > distancia_B and distancia_A > distancia_C:
    print("O ponto A está mais afastado da origem.")
elif distancia_B > distancia_A and distancia_B > distancia_C:
    print("O ponto B está mais afastado da origem.")
elif distancia_C > distancia_A and distancia_C > distancia_B:
    print("O ponto C está mais afastado da origem.")
else:
    print("Dois ou mais pontos estão à mesma distância da origem.")

# Agora, criamos um retângulo utilizando os pontos A e B
class Retangulo:
    def __init__(self, ponto1, ponto2):
        self.ponto1 = ponto1
        self.ponto2 = ponto2

    def base(self):
        return abs(self.ponto2.x - self.ponto1.x)

    def altura(self):
        return abs(self.ponto2.y - self.ponto1.y)

    def area(self):
        return self.base() * self.altura()

# Criamos um retângulo com os pontos A e B
retangulo_AB = Retangulo(ponto_A, ponto_B)

# Consultamos a base, altura e área do retângulo
print(f"Base do retângulo AB: {retangulo_AB.base()}")
print(f"Altura do retângulo AB: {retangulo_AB.altura()}")
print(f"Área do retângulo AB: {retangulo_AB.area()}")
````

````python
# Fazer o exercício aqui

# Define a dictionary to map attribute names to units
attribute_units = {
    "velocidade": "km/h",
    "cilindrada": "L",
    "carga": "kg",
    "tipo": "",  # Empty string for the "tipo" attribute since it doesn't require a unit
    "cor": "",   # Empty string for the "cor" attribute since it doesn't require a unit
    "rodas": ""  # Empty string for the "rodas" attribute since it doesn't require a unit
}

class Veiculo:
    """
    Classe que representa um veículo.
    
    Argumentos:
        cor (str): Cor do veículo.
        rodas (int): Número de rodas do veículo.

    Atributos:
        cor (str): Cor do veículo.
        rodas (int): Número de rodas do veículo.
    
    Métodos:
        __str__: Retorna a informação do veículo.
    """
    def __init__(self, cor, rodas):
        """
        Construtor da classe Veículo.

        Args:
            cor (str): Cor do veículo.
            rodas (int): Número de rodas do veículo.
        """
        self.cor = cor
        self.rodas = rodas
        
    def __str__(self):
        """
        Método que retorna a informação do veículo.

        Returns:
            str: String com a informação do veículo.
        """
        return f"cor: {self.cor}, rodas: {self.rodas}"

class Carro(Veiculo):
    """
    Classe que representa um carro.
    
    Argumentos:
        cor (str): Cor do carro.
        rodas (int): Número de rodas do carro.
        velocidade (int): Velocidade máxima do carro.
        cilindrada (float): Cilindrada do carro.
        
    Atributos:
        cor (str): Cor do carro.
        rodas (int): Número de rodas do carro.
        velocidade (int): Velocidade máxima do carro.
        cilindrada (float): Cilindrada do carro.
    
    Métodos:
        __str__: Retorna a informação do carro.
    """
    def __init__(self, cor, rodas, velocidade, cilindrada):
        """
        Construtor da classe Carro.

        Args:
            cor (str): Cor do carro.
            rodas (int): Número de rodas do carro.
            velocidade (int): Velocidade máxima do carro.
            cilindrada (float): Cilindrada do carro.
        """
        super().__init__(cor, rodas)
        self.velocidade = velocidade
        self.cilindrada = cilindrada
        
    def __str__(self):
        """
        Método que retorna a informação do carro.

        Returns:
            str: String com a informação do carro.
        """
        return f"{super().__str__()}, velocidade: {self.velocidade}, cilindrada: {self.cilindrada}"

class Camião(Carro):
    """
    Classe que representa um camião.
    
    Argumentos:
        cor (str): Cor do camião.
        rodas (int): Número de rodas do camião.
        velocidade (int): Velocidade máxima do camião.
        cilindrada (float): Cilindrada do camião.
        carga (float): Capacidade de carga do camião.
        
    Atributos:
        cor (str): Cor do camião.
        rodas (int): Número de rodas do camião.
        velocidade (int): Velocidade máxima do camião.
        cilindrada (float): Cilindrada do camião.
        carga (float): Capacidade de carga do camião.
    
    Métodos:
        __str__: Retorna a informação do camião.
    """
    def __init__(self, cor, rodas, velocidade, cilindrada, carga):
        """
        Construtor da classe Camião.

        Args:
            cor (str): Cor do camião.
            rodas (int): Número de rodas do camião.
            velocidade (int): Velocidade máxima do camião.
            cilindrada (float): Cilindrada do camião.
            carga (float): Capacidade de carga do camião.
        """
        super().__init__(cor, rodas, velocidade, cilindrada)
        self.carga = carga
        
    def __str__(self):
        """
        Método que retorna a informação do camião.

        Returns:
            str: String com a informação do camião.
        """
        return f"{super().__str__()}, carga: {self.carga}"

class Bicicleta(Veiculo):
    """
    Classe que representa uma bicicleta.
    
    Argumentos:
        cor (str): Cor da bicicleta.
        rodas (int): Número de rodas da bicicleta.
        tipo (str): Tipo da bicicleta (urbano ou desportivo).
        
    Atributos:
        cor (str): Cor da bicicleta.
        rodas (int): Número de rodas da bicicleta.
        tipo (str): Tipo da bicicleta (urbano ou desportivo).
    
    Métodos:
        __str__: Retorna a informação da bicicleta.
    """
    def __init__(self, cor, rodas, tipo):
        """
        Construtor da classe Bicicleta.

        Args:
            cor (str): Cor da bicicleta.
            rodas (int): Número de rodas da bicicleta.
            tipo (str): Tipo da bicicleta (urbano ou desportivo).
        """
        super().__init__(cor, rodas)
        self.tipo = tipo
        
    def __str__(self):
        """
        Método que retorna a informação da bicicleta.

        Returns:
            str: String com a informação da bicicleta.
        """
        return f"{super().__str__()}, tipo: {self.tipo}"

class Moto(Bicicleta):
    """
    Classe que representa uma moto.
    
    Argumentos:
        cor (str): Cor da moto.
        rodas (int): Número de rodas da moto.
        tipo (str): Tipo da moto (urbano ou desportivo).
        velocidade (int): Velocidade máxima da moto.
        cilindrada (float): Cilindrada da moto.
        
    Atributos:
        cor (str): Cor da moto.
        rodas (int): Número de rodas da moto.
        tipo (str): Tipo da moto (urbano ou desportivo).
        velocidade (int): Velocidade máxima da moto.
        cilindrada (float): Cilindrada da moto.
    
    Métodos:
        __str__: Retorna a informação da moto.
    """
    def __init__(self, cor, rodas, tipo, velocidade, cilindrada):
        """
        Construtor da classe Moto.

        Args:
            cor (str): Cor da moto.
            rodas (int): Número de rodas da moto.
            tipo (str): Tipo da moto (urbano ou desportivo).
            velocidade (int): Velocidade máxima da moto.
            cilindrada (float): Cilindrada da moto.
        """
        super().__init__(cor, rodas, tipo)
        self.velocidade = velocidade
        self.cilindrada = cilindrada
        
    def __str__(self):
        """
        Método que retorna a informação da moto.

        Returns:
            str: String com a informação da moto.
        """
        return f"{super().__str__()}, velocidade: {self.velocidade}, cilindrada: {self.cilindrada}"

# Função para catalogar os veículos
def catalogar(veiculos, rodas=None):
    """
    Função que cataloga e exibe informações dos veículos.

    Args:
        veiculos (list): Lista de objetos de veículos.
        rodas (int, optional): Número de rodas para filtrar os veículos. Defaults to None.
    """
    try:
        # Validate input for rodas
        if rodas is not None and not isinstance(rodas, int):
            raise TypeError("O argumento 'rodas' deve ser um número inteiro ou None.")
        
        # Filtrar os veículos com base no número de rodas, se o argumento 'rodas' for fornecido
        veiculos_filtrados = veiculos if rodas is None else [veiculo for veiculo in veiculos if veiculo.rodas == rodas]
        
        # Se o argumento 'rodas' foi fornecido, exibir a mensagem correspondente
        if rodas is not None:
            print(f"Encontrados {len(veiculos_filtrados)} veículos com {rodas} rodas:")
        
        # Exibir os veículos filtrados
        for veiculo in veiculos_filtrados:
            # Obtém o nome da classe do objeto
            nome_classe = veiculo.__class__.__name__
            # Converte o objeto em uma string para exibir os atributos
            atributos = str(veiculo)
            
            # Print the class name
            print(f"Classe: {nome_classe}")
                   
            # Print each attribute with its unit
            for attribute_line in atributos.split(', '):
                parts = attribute_line.split(": ", 1)  # Split at the first colon
                if len(parts) == 2:
                    attribute_name, attribute_value = parts
                    unit = attribute_units.get(attribute_name, "")
                    if unit:
                        print(f"\t{attribute_name}: {attribute_value} {unit}")
                    else:
                        print(f"\t{attribute_name}: {attribute_value}")
            print()
    
    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Função para filtrar os veículos por tipo
def filtro(veiculos, nome_tipo, attribute_units=attribute_units):
    """
    Função que filtra os veículos por tipo de veículo e exibe suas informações com unidades.

    Args:
        veiculos (list): Lista de objetos de veículos.
        nome_tipo (str): Nome do tipo de veículo a ser filtrado.
        attribute_units (dict): Dicionário que mapeia nomes de atributos para unidades.

    Returns:
        list: Lista dos veículos do tipo especificado.
    """
    try:
        # Validate input for nome_tipo
        if not isinstance(nome_tipo, str):
            raise TypeError("O argumento 'nome_tipo' deve ser uma string.")
        
        veiculos_filtrados = [veiculo for veiculo in veiculos if isinstance(veiculo, eval(nome_tipo))]
        
        print(f"Veículos do tipo {nome_tipo}:")
        
        for veiculo in veiculos_filtrados:
            for attribute_name, attribute_value in veiculo.__dict__.items():
                unit = attribute_units.get(attribute_name, "")
                if unit:
                    print(f"\t{attribute_name}: {attribute_value} {unit}")
                else:
                    print(f"\t{attribute_name}: {attribute_value}")
            print()
        
        return veiculos_filtrados
    
    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Criar objetos de cada subclasse de Veiculo
carro1 = Carro("Preto", 4, 200, 2.0)
carro2 = Carro("Branco", 4, 180, 1.5)
camião1 = Camião("Verde", 6, 120, 5.0, 8000)
camião2 = Camião("Amarelo", 8, 100, 7.0, 12000)
bicicleta1 = Bicicleta("Prata", 2, "urbano")
bicicleta2 = Bicicleta("Roxo", 2, "desportivo")
moto1 = Moto("Preta", 2, "urbano", 250, 1.8)
moto2 = Moto("Azul", 2, "desportivo", 300, 2.2)

# Criar a lista de veículos
veiculos = [carro1, carro2, camião1, camião2, bicicleta1, bicicleta2, moto1, moto2]

# Chamar a função catalogar com a lista de veículos e testar com diferentes valores de rodas
#catalogar(veiculos)  # Mostrar todos os veículos

# Chamar a função filtro para filtrar carros
lista_carros = filtro(veiculos, "Carro")
print()
lista_carros = filtro(veiculos, "Moto")
````
